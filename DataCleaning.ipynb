```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("HeirloomSeeds.csv",index_col=0)

df.rename({"Plant Information":"item9","Growing Tips":"item10","SKU":"item11"},axis=1,inplace=True)
```

```python
columns = ["LifeCycle","Sun","Sprouts","Seed Depth","Temp","Spacing","Scientific Name","Frost Hardy","SKU","Growing Tips", "Plant Information"]
df[columns] = "Missing" 
```

```python
df[:1]
```

```python
Sun_Pattern = '|'.join(["Hours","hours","Days","Sun","Minutes","minutes"])
df['LifeCycle'] = np.where(~df.item0.str.contains(Sun_Pattern), df.item0, "Missing")
df['item0'] = np.where(df.item0.str.contains(Sun_Pattern), df.item0, "Missing")

```

```python
df = df.replace(np.NaN,"Missing")
```

```python
pattern1 = '|'.join(["Hours","hours","Days","Sun","Minutes","minutes"])
import re
for i in range(df.shape[0]):
  for j in range(0,11):  

    #Sun
    if (re.search(pattern1,df.iloc[i,j]) is not None) & ("Sprouts" not in df.iloc[i,j]) & (len(df.iloc[i,j]) < 20):
      df["Sun"][i] = df.iloc[i,j]
      df.iloc[i,j] = "Missing"

    #Sprouts Column
    if ("Sprouts in" in df.iloc[i,j]):
      df["Sprouts"][i] = df.iloc[i,j]
      df.iloc[i,j] = "Missing"

    #Seed Depth
    if "Seed Depth" in df.iloc[i,j]:   
      df["Seed Depth"][i] = df.iloc[i,j]
      df.iloc[i,j] = "Missing"
    #Temperature
    if "Ideal Temperature" in df.iloc[i,j]:   
      df["Temp"][i] = df.iloc[i,j]
      df.iloc[i,j] = "Missing"
    #Spacing
    if "Plant Spacing" in df.iloc[i,j]:   
      df["Spacing"][i] = df.iloc[i,j]
      df.iloc[i,j] = "Missing"
    #Frost Hardy
    if "Frost Hardy" in df.iloc[i,j]:
      df["Frost Hardy"][i] = df.iloc[i,j]
      df.iloc[i,j] = "Missing"

    #Growing Tips
    if "Growing Tips" in df.iloc[i,j]:
      df["Growing Tips"][i] = df.iloc[i,j]
      df.iloc[i,j] = "Missing" 

    #Sku
    if len(df.iloc[i,j]) == 5:
      df["SKU"][i] = df.iloc[i,j]
      df.iloc[i,j] = "Missing"   


    if 1 < len(df.iloc[i,j].split(" ")) < 5:
      df["Scientific Name"][i] = df.iloc[i,j]
      df.iloc[i,j] = "Missing"

  



for i in range(df.shape[0]):
  for j in range(11):
    if df.iloc[i,j] != "Missing":
      df["Plant Information"][i] = df.iloc[i,j]
      break


   



  
```

```python
unique_LifeCycle_values = set(pd.Series(df.LifeCycle.value_counts().index).str.strip().replace("Annual/Biennial","Annual or Biennial").to_list())
for index,info in enumerate(df["Plant Information"]):
  for life_cycle in unique_LifeCycle_values:
    if life_cycle in info:
      df.LifeCycle[index] = life_cycle
      break
```

```python
dtm = []
for item in df["Plant Information"].to_list():
  string = item.split(". ")[0]
  if (re.search("days|Days",string) is not None) & (len(string) < 20):
    dtm.append(string)
  else:
    dtm.append("Missing")

df["Days to Maturity"] = dtm
```

```python
columns_to_drop = ["item" + str(i) for i in range(12)]
df.drop(columns_to_drop,axis=1,inplace=True)
```

```python
df
```

```python
df["Seed Depth"] = df["Seed Depth"].str.replace("Seed Depth: ","")
df["Temp"] = df["Temp"].str.replace("Ideal Temperature: ","")
df["Spacing"] = df["Spacing"].str.replace("Plant Spacing: ","")
df["Frost Hardy"] = df["Frost Hardy"].str.replace("Frost Hardy: ","")
df["Growing Tips"] = df["Growing Tips"].str.replace("Growing Tips: ","")
```

```python
for col in df:
  df[col] = df[col].str.strip()

df["Frost Hardy"] = df["Frost Hardy"].str.capitalize()

df.Sun = df.Sun.str.replace("hrs","hours")

df.loc[df.Sun.str.contains("Days"),"Sun"] = "Full Sun"
df.Sprouts = df.Sprouts.str.replace("Full Sun ","")

df.Temp = df.Temp.str.replace(" Degrees F","F")

for index,item in enumerate(df["Scientific Name"]):
  if "days" in item:
    df["Scientific Name"].iloc[index] = "Missing"
```

```python
df.to_csv("HeirloomsFull.csv")
```
